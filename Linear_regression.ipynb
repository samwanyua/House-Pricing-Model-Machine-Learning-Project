```python

X=dataset.iloc[:, :-1] # select all rows (:) and all columns except the last one (:-1).
# target variable
y= dataset.iloc[:, -1] # selects all rows (:) and only the last column (-1).
```

```python
X.head()
```

```python
y
```

#### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

#### Standardize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train) # The training data (X_train) is scaled so that each feature has a mean of 0 and a standard deviation of 1.
```

```python
X_test = scaler.transform(X_test) # The test data is normalized using the same scaling parameters as the training data, maintaining consistency between training and testing datasets.
```

```python
import pickle
pickle.dump(scaler, open('scaling.pkl', 'wb')) # wb -> write binary
```

```python
X_train
```

```python
X_test
```

### Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
regression.coef_ # print coefficients and intercept
```

```python
regression.intercept_
```

```python
#params on which the model has been trained
regression.get_params()
```

```python
# prediction with test data
reg_predict = regression.predict(X_test)
```

```python
reg_predict
```

### Assumption

```python
# plot a scatter plot for the prediction
plt.scatter(y_test, reg_predict)
plt.title("Actual vs Predicted House Prices")
plt.xlabel("Actual House Price")
plt.ylabel("Predicted House Price")
plt.show()
```

```python
# Residuals
residuals = y_test - reg_predict
residuals
```

##### Plot residuals

```python
sns.displot(residuals, kind="kde")
plt.title("Kernel Density Estimate of Residuals")
plt.show()
```

##### Scatter plot with respect to prediction and residuals

```python
plt.scatter(reg_predict, residuals)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(f"Mean absolute error - {mean_absolute_error(y_test, reg_predict)}")
print(f"Mean squared error - {mean_squared_error(y_test, reg_predict)}")
print(f"Root Mean Squared Error - {np.sqrt(mean_squared_error(y_test, reg_predict))}")
```

### Conclusions
##### MAE of approximately 0.527 indicates that, on average, the model's predictions are off by about 0.527 units of the target variable (house prices).

##### MSE of approximately 0.531 shows that there are some larger errors that are being squared, which increases the overall error metric.

##### RMSE of approximately 0.728 provides a more intuitive measure of error in the same units as the target variable, indicating that the average prediction error is around 0.728 units of house prices.

### R square and adjusted R square
##### Formula

##### R^2 = 1 - SSR/SST

##### R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
from sklearn.metrics import r2_score
score= r2_score(y_test, reg_predict)
print(f"Score: {score}")
```

##### This score indicates that approximately 59.58% of the variability in the target variable (y_test, house prices) is explained by the regression model (reg_predict).

#### In other words, the model captures around 59.58% of the variance in house prices based on the features used in the model

### Adjusted R2 = 1 - [(1-R2) * (n-1)/(n-k-1)]
#### where:

#### R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
print(f"Adjusted R_squared : {1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)}")
```
